### Import Libraries

````python
# Import Libraries
import os
import json
import sqlite3
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset
!git clone https://github.com/PhonePe/pulse.git
````

````python
#Create SQL Database
conn = sqlite3.connect('phonepe.db')
cursor = conn.cursor()
````

````python
#Create Tables
cursor.execute("""
CREATE TABLE IF NOT EXISTS aggregated_transaction (
    state TEXT,
    year INT,
    quarter INT,
    transaction_type TEXT,
    transaction_count INT,
    transaction_amount REAL
)
""")
````

````python
#agrregated user table
cursor.execute("""
CREATE TABLE IF NOT EXISTS aggregated_user (
   state TEXT,
   year INT,
   quarter INT,
   brands TEXT,
   transaction_count INT,
   percentage REAL
)
""")
````

````python
#Aggregated Insurance Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS aggregated_insurance (
   state TEXT,
   year INT,
   quarter INT,
   insurance_type TEXT,
   transaction_count INT,
   transaction_amount REAL
)
""")
````

````python
#Extract JSON → Insert into SQL
path = "/content/pulse/data/aggregated/transaction/country/india/state/"

for state in os.listdir(path):
    state_path = path + state + "/"

    for year in os.listdir(state_path):
        year_path = state_path + year + "/"

        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = int(file.strip(".json"))

                with open(year_path + file) as f:
                    data = json.load(f)

                    try:
                        for item in data["data"]["transactionData"]:
                            t_type = item["name"]

                            for payment in item["paymentInstruments"]:
                                count = payment["count"]
                                amount = payment["amount"]

                                cursor.execute("""
                                INSERT INTO aggregated_transaction
                                VALUES (?, ?, ?, ?, ?, ?)
                                """, (state, int(year), quarter, t_type, count, amount))
                    except:
                        pass
````

````python
base_path = "pulse/data/aggregated/insurance/country/india/state/"

for state in os.listdir(base_path):
    state_path = base_path + state + "/"

    for year in os.listdir(state_path):
        year_path = state_path + year + "/"

        for file in os.listdir(year_path):
            if file.endswith(".json"):
                quarter = int(file.strip(".json"))

                with open(year_path + file) as f:
                    data = json.load(f)

                    try:
                        for item in data["data"]["transactionData"]:
                            i_type = item["name"]

                            for ins in item["paymentInstruments"]:
                                count = ins["count"]
                                amount = ins["amount"]

                                cursor.execute("""
                                INSERT INTO aggregated_insurance
                                VALUES (?, ?, ?, ?, ?, ?)
                                """, (state, int(year), quarter, i_type, count, amount))
                    except:
                        pass
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df_trnx = pd.read_sql_query("SELECT * FROM aggregated_transaction", conn)
df_user = pd.read_sql_query("SELECT * FROM aggregated_user", conn)
df_inrns = pd.read_sql_query("SELECT * FROM aggregated_insurance", conn)
````

````python
#handiling missing values
df_trnx.dropna(inplace=True)
df_user.dropna(inplace=True)
df_inrns.dropna(inplace=True)
````

````python
df_trnx = df_trnx.drop_duplicates()
df_user = df_user.drop_duplicates()
df_inrns = df_inrns.drop_duplicates()
````

````python
df_trnx['state'] = df_trnx['state'].str.lower().str.replace('-', ' ')
df_user['state'] = df_user['state'].str.lower().str.replace('-', ' ')
df_inrns['state'] = df_inrns['state'].str.lower().str.replace('-', ' ')
````

````python
#fix data types
df_trnx['year'] = df_trnx['year'].astype(int)
df_trnx['quarter'] = df_trnx['quarter'].astype(int)
df_user['year'] = df_user['year'].astype(int)
df_user['quarter'] = df_user['quarter'].astype(int)
df_inrns['year'] = df_inrns['year'].astype(int)
df_inrns['quarter'] = df_inrns['quarter'].astype(int)
````

````python
#invalid values
df_trnx = df_trnx[df_trnx['transaction_count'] >= 0]
df_trnx = df_trnx[df_trnx['transaction_amount'] >= 0]
df_inrns = df_inrns[df_inrns['transaction_count'] >= 0]
df_inrns = df_inrns[df_inrns['transaction_amount'] >= 0]
````

````python
#saving back to sql
df_trnx.to_sql('clean_aggregated_transaction', conn, if_exists='replace', index=False)
df_user.to_sql('clean_aggregated_user', conn, if_exists='replace', index=False)
df_inrns.to_sql('clean_aggregated_insurance', conn, if_exists='replace', index=False)
````

### What all manipulations have you done and insights you found?

Removed missing/null values from all tables
Eliminated duplicate records
Standardized state names (converted to lowercase and removed special characters)
Ensured correct data types for year, quarter, count, and amount columns
Filtered out invalid records (zero or negative transaction/insurance values)
